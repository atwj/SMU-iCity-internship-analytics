```python
#imports
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#plot
# Create the figure and configure
fig = plt.figure(figsize=(8,8))
# Add subplot of polar projection
ax1 = fig.add_subplot(111, projection='polar')
# Add 24 xtick points of (0,23) representing the hours in a day
ax1.set_xticks(np.linspace(0,23,24,endpoint=True, dtype=np.int32) / 24 * (2 * np.pi))
# Generate xtick labels from [0,23]
xticklabels = [str(x) for x in range (0,24)]
# Set xtick labels
ax1.set_xticklabels(xticklabels)
# Set 0 degrees to 'North'
ax1.set_theta_zero_location('N')
# Set values to increment clockwise
ax1.set_theta_direction(-1)
ax1.plot ([0.1,0.2,0.3,0.5],'x')

plt.show()
```
